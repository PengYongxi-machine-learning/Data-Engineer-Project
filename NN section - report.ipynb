#### Model Architecture and Methodology

The predictive methodology employs a split-architecture strategy, developing independent feed-forward neural networks for specific vehicle types and distinct driving environments—specifically targeting City and Highway fuel consumption. By filtering the data to one vehicle type at a time, this granular approach allows the networks to capture segment-specific non-linear relationships that a singular global model might overlook. The core neural network is implemented utilizing PyTorch and features two hidden layers. To ensure training stability and mitigate overfitting, batch normalization and dropout regularization layers are sequentially applied after each hidden linear transformation. The Exponential Linear Unit (ELU) is established as the global activation function within the network structure.

#### Data Preprocessing

Prior to training, the dataset undergoes rigorous preprocessing. Independent target vectors are isolated for City and Highway fuel consumption metrics. To prevent unstable estimates and high error variance, vehicle categories containing fewer than 50 samples are deliberately excluded from the pipeline. Categorical features are one-hot encoded, while numerical feature matrices are standardized utilizing a StandardScaler. Crucially, this scaler is fitted exclusively on the training partition to avoid data leakage. A reproducible 80/20 train-test split is applied using a fixed random state, guaranteeing fair performance comparisons across various tuning and training iterations.

#### Hyperparameter Optimization

To maximize model efficacy, the architecture integrates automated hyperparameter tuning through the Optuna framework. For each specific vehicle type and target environment, an optimization study minimizes the validation loss by systematically exploring a defined parameter search space. The objective function optimizes network capacity and learning dynamics by tuning the number of units in the first and second hidden layers, dropout probabilities (ranging from 0.1 to 0.5), Adam optimizer learning rates (sampled log-uniformly), and data loader batch sizes (32, 64, or 128).

#### Training Protocol and Evaluation

The final model training phase employs the tuned hyperparameters and utilizes the Adam optimizer to minimize Mean Squared Error (MSE) loss across tensor mini-batches. A robust early stopping mechanism is integrated into the training loop. This mechanism actively tracks validation loss against a predefined patience threshold, halting the training process when generalization improvements cease and subsequently restoring the model's best-performing state dictionary. Final model evaluation on the held-out test data is quantified utilizing Root Mean Squared Error (RMSE) and the coefficient of determination ($R^2$). Furthermore, the evaluation pipeline generates side-by-side diagnostic scatter plots comparing predicted versus actual fuel consumption. These plots incorporate a perfect-prediction diagonal line to visually assess model accuracy, systematically diagnose prediction bias, and observe error variance.
