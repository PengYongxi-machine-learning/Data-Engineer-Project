#### Modules

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from pathlib import Path
```

#### Basic information of the raw dataset

```python
#Import libraries and upload data (We ignore the effect of the differences in car models and brands)
df = pd.read_csv('Raw dataset.csv')
df = df.drop(columns = ['Make', 'Model'])
print(df.head())

#Lets look at the number of rows and columns and the data type for each feature
print(df.info())
```

#### Check missing data and drop duplicates

```python
#Looking for duplicates and dropping any duplicates (duplicates are dropped)
print(df.duplicated().sum())
df = df.drop_duplicates()

# Let's look at the missing values by feature (no missing data)
df.isnull().sum()
```

#### Visualize the distribution of the dataset

```python
All_cols = df.columns
for col in All_cols:
    sns.histplot(df[col])
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=90)
    plt.show()
```

#### Whisker box plots

```python
#Using the boxplot function via seaborn
for col in All_cols:
    sns.boxplot(df[col])
    plt.xticks(rotation=90) 
    plt.show()
```

#### Even though there many points being identified as outliers by the whisker box plots, we did not remove any of them because:
1. None of the data is normally distributed, thus the outliers can not be founded using whisker box plots.

2. The identified outliers do not have abnormal numerical values (data anomaly), thus hard to conclude whether they are actual outliers or not

```python
# Output cleaned dataset
output_path = Path("Clean dataset.csv")
df.to_csv(output_path, index=False)
```
